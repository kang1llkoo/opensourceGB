```python
import random

def yut_throw():
    result = random.choices(["앞", "뒤"], k=4)
    result_str = ' '.join(result)
    if result.count("뒤") == 1:
        return result_str + " (도)"
    elif result.count("뒤") == 2:
        return result_str + " (개)"
    elif result.count("뒤") == 3:
        return result_str + " (걸)"
    elif result.count("뒤") == 4:
        return result_str + " (윷)"
    else:
        return result_str + " (모)"

def main():
    player_a_pos = 0
    player_b_pos = 0

    while player_a_pos < 30 and player_b_pos < 30:
        input("A의 차례입니다. 가위바위보를 시작하려면 엔터를 누르세요.")
        a_choice = random.choice(["가위", "바위", "보"])
        print(f"A의 선택: {a_choice}")

        input("B의 차례입니다. 가위바위보를 시작하려면 엔터를 누르세요.")
        b_choice = random.choice(["가위", "바위", "보"])
        print(f"B의 선택: {b_choice}")

        if a_choice == b_choice:
            print("무승부! 다시 가위바위보를 진행합니다.")
            continue
        elif (
            (a_choice == "가위" and b_choice == "보") or
            (a_choice == "바위" and b_choice == "가위") or
            (a_choice == "보" and b_choice == "바위")
        ):
            print("A가 이겼습니다!")
            player_a_pos = 0
        else:
            print("B가 이겼습니다!")
            player_b_pos = 0

        while player_a_pos < 30 and player_b_pos < 30:
            while True:
                input("A의 차례입니다. 윷을 던지려면 엔터를 누르세요.")
                yut_result = yut_throw()
                print(f"A의 윷 결과: {yut_result}")
                if '윷' in yut_result or '모' in yut_result:
                    print("윷이나 모가 나왔습니다. 한 번 더 던지십시오.")
                else:
                    break

            move_steps = yut_result.count("도") + 2 * yut_result.count("개") + 3 * yut_result.count("걸") + 4 * yut_result.count("윷") + 5 * yut_result.count("모")
            if '모' in yut_result:
                move_steps += 5
            player_a_pos += move_steps

            if player_a_pos == 5:
                print("말이 5번에 도착하고 다음 차례에 1번 루트로 이동합니다.")
                player_a_pos = 0
            elif player_a_pos == 22:
                print("말이 22번에 도착하고 다음 차례에 4번 루트로 이동합니다.")
                player_a_pos = 28
            elif player_a_pos == 10:
                print("말이 10번에 도착하고 다음 차례에 2번 루트로 이동합니다.")
                player_a_pos = 25

            if player_a_pos > 30:
                player_a_pos = 30

            print(f"A의 위치: {player_a_pos}")

            if player_a_pos >= 30:
                break

            while True:
                input("B의 차례입니다. 윷을 던지려면 엔터를 누르세요.")
                yut_result = yut_throw()
                print(f"B의 윷 결과: {yut_result}")
                if '윷' in yut_result or '모' in yut_result:
                    print("윷이나 모가 나왔습니다. 한 번 더 던지십시오.")
                else:
                    break

            move_steps = yut_result.count("도") + 2 * yut_result.count("개") + 3 * yut_result.count("걸") + 4 * yut_result.count("윷") + 5 * yut_result.count("모")
            if '모' in yut_result:
                move_steps += 5
            player_b_pos += move_steps

            if player_b_pos == 5:
                print("말이 5번에 도착하고 다음 차례에 1번 루트로 이동합니다.")
                player_b_pos = 0
            elif player_b_pos == 22:
                print("말이 22번에 도착하고 다음 차례에 4번 루트로 이동합니다.")
                player_b_pos = 28
            elif player_b_pos == 10:
                print("말이 10번에 도착하고 다음 차례에 2번 루트로 이동합니다.")
                player_b_pos = 25

            if player_b_pos > 30:
                player_b_pos = 30

            print(f"B의 위치: {player_b_pos}")

            if player_b_pos >= 30:
                break

    if player_a_pos >= 30:
        print("플레이어 A가 이겼습니다! 축하합니다!")
    else:
        print("플레이어 B가 이겼습니다! 축하합니다!")

if __name__ == "__main__":
    main()
```

```python
import random

def yut_throw():
    result = random.choices(["앞", "뒤"], k=4)
    result_str = ' '.join(result)
    if result.count("뒤") == 1:
        return result_str + " (도)"
    elif result.count("뒤") == 2:
        return result_str + " (개)"
    elif result.count("뒤") == 3:
        return result_str + " (걸)"
    elif result.count("뒤") == 4:
        return result_str + " (윷)"
    else:
        return result_str + " (모)"

def main():
    player_a_pos = 0
    player_b_pos = 0
    current_player = None

    while player_a_pos < 30 and player_b_pos < 30:
        input("A의 차례입니다. 가위바위보를 시작하려면 엔터를 누르세요.")
        a_choice = random.choice(["가위", "바위", "보"])
        print(f"A의 선택: {a_choice}")

        input("B의 차례입니다. 가위바위보를 시작하려면 엔터를 누르세요.")
        b_choice = random.choice(["가위", "바위", "보"])
        print(f"B의 선택: {b_choice}")

        if a_choice == b_choice:
            print("무승부! 다시 가위바위보를 진행합니다.")
            continue
        elif (
            (a_choice == "가위" and b_choice == "보") or
            (a_choice == "바위" and b_choice == "가위") or
            (a_choice == "보" and b_choice == "바위")
        ):
            print("A가 이겼습니다!")
            current_player = 'A'
            player_a_pos = 0
        else:
            print("B가 이겼습니다!")
            current_player = 'B'
            player_b_pos = 0

        while player_a_pos < 30 and player_b_pos < 30:
            while True:
                input(f"{current_player}의 차례입니다. 윷을 던지려면 엔터를 누르세요.")
                yut_result = yut_throw()
                print(f"{current_player}의 윷 결과: {yut_result}")
                if '윷' in yut_result:
                    print("윷이 나왔습니다. 한 번 더 던지십시오. ")
                    input("엔터를 눌러 한 번 더 던지세요. ")
                    yut_result += ' ' + yut_throw()  # 한 번 더 던짐
                    print(f"{current_player}의 추가 윷 결과: {yut_result.split()[-1]}")
                elif '모' in yut_result:
                    print("모가 나왔습니다. 한 번 더 던지십시오.")
                    input("엔터를 눌러 한 번 더 던지세요.")
                    yut_result += ' ' + yut_throw()  # 한 번 더 던짐
                    print(f"{current_player}의 추가 윷 결과: {yut_result.split()[-1]}")
                break

            # Check if the players' pieces are on the same position after throwing yut
            if player_a_pos == player_b_pos and player_a_pos != 0:
                print(f"{current_player}가 상대방의 말을 잡았습니다!")

                # Reset the opponent's position to the starting point (0)
                if current_player == 'A':
                    player_b_pos = 0
                else:
                    player_a_pos = 0

                print(f"A의 위치: {player_a_pos}, B의 위치: {player_b_pos}")

            move_steps = 0
            if '도' in yut_result:
                move_steps += yut_result.count('도')
            if '개' in yut_result:
                move_steps += yut_result.count('개') * 2
            if '걸' in yut_result:
                move_steps += yut_result.count('걸') * 3
            if '윷' in yut_result:
                move_steps += yut_result.count('윷') * 4
            if '모' in yut_result:
                move_steps += yut_result.count('모') * 5

            if current_player == 'A':
                player_a_pos += move_steps

                if player_a_pos == 5:
                    print("말이 5번에 도착하고 다음 차례에 1번 루트로 이동합니다.")
                    player_a_pos = 20
                    print(f"{current_player}의 위치: {player_a_pos}")
                elif player_a_pos == 22 or player_a_pos == 27:
                    print("말이 22번에 도착하고 다음 차례에 1번 루트로 이동합니다.")
                    player_a_pos = 22
                    print(f"{current_player}의 위치: {player_a_pos}")
                elif player_a_pos == 10:
                    print("말이 10번에 도착하고 다음 차례에 2번 루트로 이동합니다.")
                    player_a_pos = 25
                    print(f"{current_player}의 위치: {player_a_pos}")

                if player_a_pos > 30:
                    player_a_pos = 30
            else:
                player_b_pos += move_steps

                if player_b_pos == 5:
                    print("말이 5번에 도착하고 다음 차례에 1번 루트로 이동합니다.")
                    player_b_pos = 20
                    print(f"{current_player}의 위치: {player_b_pos}")
                elif player_b_pos == 22 or player_b_pos == 27:
                    print("말이 22번에 도착하고 다음 차례에 1번 루트로 이동합니다.")
                    player_b_pos = 22
                    print(f"{current_player}의 위치: {player_b_pos}")
                elif player_b_pos == 10:
                    print("말이 10번에 도착하고 다음 차례에 2번 루트로 이동합니다.")
                    player_b_pos = 25
                    print(f"{current_player}의 위치: {player_b_pos}")

                if player_b_pos > 30:
                    player_b_pos = 30

            print(f"{current_player}의 위치: {player_a_pos if current_player == 'A' else player_b_pos}")

            if player_a_pos >= 30 or player_b_pos >= 30:
                break

            # Check if the players' pieces are on the same position after moving
            if player_a_pos == player_b_pos and player_a_pos != 0:
                print(f"{current_player}가 상대방의 말을 잡았습니다!")

                # Reset the opponent's position to the starting point (0)
                if current_player == 'A':
                    player_b_pos = 0
                else:
                    player_a_pos = 0

                print(f"A의 위치: {player_a_pos}, B의 위치: {player_b_pos}")

            current_player = 'B' if current_player == 'A' else 'A'

    if player_a_pos >= 30:
        print("플레이어 A가 이겼습니다! 축하합니다!")
    else:
        print("플레이어 B가 이겼습니다! 축하합니다!")

if __name__ == "__main__":
    main()
```
