```python
import random

def yut_throw():
    result = random.choices(["앞"] * 4 + ["뒤"] * 3 + ["뒤-"], k=4)
    result_str = ' '.join(result)
    
    if result.count("뒤-") > 1:
        result_str = result_str.replace("뒤-", "뒤", 2)
    
    if result.count("뒤") == 1 and result.count("앞") == 3:
        return result_str + " (도)"
    elif result.count("앞") == 3 and result.count("뒤-") == 1:
        return result_str + " (빽도)"
    elif result.count("앞") == 2:
        return result_str + " (개)"
    elif result.count("앞") == 1:
        return result_str + " (걸)"
    elif result.count("앞") == 0:
        return result_str + " (윷)"
    elif result.count("앞") == 4:
        return result_str + " (모)"

def main():
    player_a_pos = 0
    player_b_pos = 0
    current_player = None

    while player_a_pos < 30 and player_b_pos < 30:
        input("A의 차례입니다. 가위바위보를 시작하려면 엔터를 누르세요.")
        a_choice = random.choice(["가위", "바위", "보"])
        print(f"A의 선택: {a_choice}")

        input("B의 차례입니다. 가위바위보를 시작하려면 엔터를 누르세요.")
        b_choice = random.choice(["가위", "바위", "보"])
        print(f"B의 선택: {b_choice}")

        if a_choice == b_choice:
            print("무승부! 다시 가위바위보를 진행합니다.")
            continue
        elif (
            (a_choice == "가위" and b_choice == "보") or
            (a_choice == "바위" and b_choice == "가위") or
            (a_choice == "보" and b_choice == "바위")
        ):
            print("A가 이겼습니다!")
            current_player = 'A'
        else:
            print("B가 이겼습니다!")
            current_player = 'B'

        while player_a_pos < 30 and player_b_pos < 30:
            while True:
                input(f"{current_player}의 차례입니다. 윷을 던지려면 엔터를 누르세요.")
                yut_result = yut_throw()
                print(f"{current_player}의 윷 결과: {yut_result}")

                # 특별한 경우에 대한 규칙 추가
                if current_player == 'A' and player_a_pos == 5:
                    if '도' in yut_result:
                        player_a_pos = 19
                    elif '빽도' in yut_result:
                        player_a_pos = 4
                    elif '개' in yut_result:
                        player_a_pos = 19
                    elif '걸' in yut_result:
                        player_a_pos = 19
                    elif '윷' in yut_result:
                        player_a_pos = 19
                    elif '모' in yut_result:
                        player_a_pos = 19

                elif current_player == 'A' and player_a_pos == 10:
                    if '도' in yut_result:
                        player_a_pos = 24
                    elif '빽도' in yut_result:
                        player_a_pos = 9
                    elif '개' in yut_result:
                        player_a_pos = 24
                    elif '걸' in yut_result:
                        player_a_pos = 24
                    elif '윷' in yut_result:
                        player_a_pos = 24
                    elif '모' in yut_result:
                        player_a_pos = 24

                elif current_player == 'A' and player_a_pos == 22:
                    if '도' in yut_result:
                        player_a_pos = 27
                    elif '빽도' in yut_result:
                        player_a_pos = 21
                    elif '개' in yut_result:
                        player_a_pos = 27
                    elif '걸' in yut_result:
                        player_a_pos = 27
                    elif '윷' in yut_result:
                        player_a_pos = 27
                    elif '모' in yut_result:
                        player_a_pos = 27
                        
                elif current_player == 'A' and player_a_pos == 24:
                    if '도' in yut_result:
                        player_a_pos = 14
                    elif '빽도' in yut_result:
                        player_a_pos = 23
                    elif '개' in yut_result:
                        player_a_pos = 14
                    elif '걸' in yut_result:
                        player_a_pos = 14
                    elif '윷' in yut_result:
                        player_a_pos = 14
                    elif '모' in yut_result:
                        player_a_pos = 14
                        
                elif current_player == 'A' and player_a_pos == 16:
                    if '도' in yut_result:
                        player_a_pos = 16
                    elif '빽도' in yut_result:
                        player_a_pos = 15
                    elif '개' in yut_result:
                        player_a_pos = 16
                    elif '걸' in yut_result:
                        player_a_pos = 16
                    elif '윷' in yut_result:
                        player_a_pos = 30
                    elif '모' in yut_result:
                        player_a_pos = 30
                        
                elif current_player == 'A' and player_a_pos == 17:
                    if '도' in yut_result:
                        player_a_pos = 17
                    elif '빽도' in yut_result:
                        player_a_pos = 16
                    elif '개' in yut_result:
                        player_a_pos = 17
                    elif '걸' in yut_result:
                        player_a_pos = 30
                    elif '윷' in yut_result:
                        player_a_pos = 30
                    elif '모' in yut_result:
                        player_a_pos = 30
                        
                elif current_player == 'A' and player_a_pos == 18:
                    if '도' in yut_result:
                        player_a_pos = 18
                    elif '빽도' in yut_result:
                        player_a_pos = 17
                    elif '개' in yut_result:
                        player_a_pos = 30
                    elif '걸' in yut_result:
                        player_a_pos = 30
                    elif '윷' in yut_result:
                        player_a_pos = 30
                    elif '모' in yut_result:
                        player_a_pos = 30
                
                elif current_player == 'A' and player_a_pos == 21:
                    if '윷' in yut_result:
                        player_a_pos = 11
                    elif '모' in yut_result:
                        player_a_pos = 11
                        
                elif current_player == 'A' and player_a_pos == 23:
                    if '윷' in yut_result:
                        player_a_pos = 13
                    elif '모' in yut_result:
                        player_a_pos = 13
                        
                elif current_player == 'A' and player_a_pos == 24:
                    if '윷' in yut_result:
                        player_a_pos = 14
                    elif '모' in yut_result:
                        player_a_pos = 14

                elif current_player == 'A' and player_a_pos == 19 and '빽도' not in yut_result:
                    player_a_pos = 30

                if current_player == 'B' and player_b_pos == 5:
                    if '도' in yut_result:
                        player_b_pos = 19
                    elif '빽도' in yut_result:
                        player_a_pos = 4
                    elif '개' in yut_result:
                        player_b_pos = 19
                    elif '걸' in yut_result:
                        player_b_pos = 19
                    elif '윷' in yut_result:
                        player_b_pos = 19
                    elif '모' in yut_result:
                        player_b_pos = 19

                elif current_player == 'B' and player_b_pos == 10:
                    if '도' in yut_result:
                        player_b_pos = 24
                    elif '빽도' in yut_result:
                        player_a_pos = 9
                    elif '개' in yut_result:
                        player_b_pos = 24
                    elif '걸' in yut_result:
                        player_b_pos = 24
                    elif '윷' in yut_result:
                        player_b_pos = 24
                    elif '모' in yut_result:
                        player_b_pos = 24

                elif current_player == 'B' and player_b_pos == 22:
                    if '도' in yut_result:
                        player_b_pos = 27
                    elif '빽도' in yut_result:
                        player_a_pos = 21
                    elif '개' in yut_result:
                        player_b_pos = 27
                    elif '걸' in yut_result:
                        player_b_pos = 27
                    elif '윷' in yut_result:
                        player_b_pos = 27
                    elif '모' in yut_result:
                        player_b_pos = 27
                        
                elif current_player == 'B' and player_b_pos == 24:
                    if '도' in yut_result:
                        player_b_pos = 14
                    elif '빽도' in yut_result:
                        player_a_pos = 23
                    elif '개' in yut_result:
                        player_b_pos = 14
                    elif '걸' in yut_result:
                        player_b_pos = 14
                    elif '윷' in yut_result:
                        player_b_pos = 14
                    elif '모' in yut_result:
                        player_b_pos = 14
                        
                elif current_player == 'B' and player_b_pos == 16:
                    if '도' in yut_result:
                        player_b_pos = 16
                    elif '빽도' in yut_result:
                        player_a_pos = 15
                    elif '개' in yut_result:
                        player_b_pos = 16
                    elif '걸' in yut_result:
                        player_b_pos = 16
                    elif '윷' in yut_result:
                        player_b_pos = 30
                    elif '모' in yut_result:
                        player_b_pos = 30
                        
                elif current_player == 'B' and player_b_pos == 17:
                    if '도' in yut_result:
                        player_b_pos = 17
                    elif '빽도' in yut_result:
                        player_a_pos = 16
                    elif '개' in yut_result:
                        player_b_pos = 17
                    elif '걸' in yut_result:
                        player_b_pos = 30
                    elif '윷' in yut_result:
                        player_b_pos = 30
                    elif '모' in yut_result:
                        player_b_pos = 30
                        
                elif current_player == 'B' and player_b_pos == 18:
                    if '도' in yut_result:
                        player_b_pos = 18
                    elif '빽도' in yut_result:
                        player_a_pos = 17
                    elif '개' in yut_result:
                        player_b_pos = 30
                    elif '걸' in yut_result:
                        player_b_pos = 30
                    elif '윷' in yut_result:
                        player_b_pos = 30
                    elif '모' in yut_result:
                        player_b_pos = 30
                        
                elif current_player == 'B' and player_b_pos == 21:
                    if '윷' in yut_result:
                        player_b_pos = 11
                    elif '모' in yut_result:
                        player_b_pos = 11
                        
                elif current_player == 'B' and player_b_pos == 23:
                    if '윷' in yut_result:
                        player_b_pos = 13
                    elif '모' in yut_result:
                        player_b_pos = 13
                        
                elif current_player == 'B' and player_b_pos == 24:
                    if '윷' in yut_result:
                        player_b_pos = 14
                    elif '모' in yut_result:
                        player_b_pos = 14

                elif current_player == 'B' and player_b_pos == 19 and '빽도' not in yut_result:
                    player_b_pos = 30
                    

                if '윷' in yut_result or '모' in yut_result:
                    yut_position_before_throw = player_a_pos if current_player == 'A' else player_b_pos
                    print(f"{current_player}의 윷이나 모가 나온 후의 위치: {yut_position_before_throw + 4 if '윷' in yut_result else yut_position_before_throw + 5}")

                    input("엔터를 눌러 한 번 더 던지세요.")
                    yut_result_extra = yut_throw()
                    print(f"{current_player}의 추가 윷 결과: {yut_result_extra}")

                    move_steps_extra = 0
                    if '도' in yut_result_extra:
                        move_steps_extra += 1
                    if '빽도' in yut_result_extra:
                        move_steps_extra -= 1
                    if '개' in yut_result_extra:
                        move_steps_extra += 2
                    if '걸' in yut_result_extra:
                        move_steps_extra += 3
                    if '윷' in yut_result_extra:
                        move_steps_extra += 4
                    if '모' in yut_result_extra:
                        move_steps_extra += 5

                    if '빽도' in yut_result_extra:
                        move_steps_extra -= 1
                        if current_player == 'A':
                            if player_a_pos > 0:
                                player_a_pos += move_steps_extra
                        elif current_player == 'B':
                            if player_b_pos > 0:
                                player_b_pos += move_steps_extra
                    else:
                        if current_player == 'A':
                            player_a_pos = yut_position_before_throw + move_steps_extra
                        elif current_player == 'B':
                            player_b_pos = yut_position_before_throw + move_steps_extra

                    player_a_pos = min(player_a_pos, 30)
                    player_b_pos = min(player_b_pos, 30)
                    
                break

            # 이동 로직 (나머지 코드는 그대로 유지됨)
            move_steps = 0
            if '도' in yut_result:
                move_steps += yut_result.count('도')
            if '빽도' in yut_result:
                if current_player == 'A':
                    if player_a_pos != 0:
                        move_steps -= yut_result.count('빽도')
                        player_a_pos -= 1  # 수정된 부분: 빽도가 나오면 플레이어 A의 위치를 -1만큼 감소
                    elif player_a_pos == 0:
                        move_steps +=0
                        player_a_pos += 0
                elif current_player == 'B':
                    if player_b_pos != 0:
                        move_steps -= yut_result.count('빽도')
                        player_b_pos -= 1
                    elif player_b_pos == 0:
                        move_steps += 0
                        player_b_pos += 0
            if '개' in yut_result:
                move_steps += yut_result.count('개') * 2
            if '걸' in yut_result:
                move_steps += yut_result.count('걸') * 3
            if '윷' in yut_result:
                move_steps += yut_result.count('윷') * 4
            if '모' in yut_result:
                move_steps += yut_result.count('모') * 5

            if current_player == 'A':
                player_a_pos += move_steps
                
            if current_player == 'B':
                player_b_pos += move_steps
                
            player_a_pos = min(player_a_pos, 30)
            player_b_pos = min(player_b_pos, 30)

            print(f"{current_player}의 위치: {player_a_pos if current_player == 'A' else player_b_pos}")

            if player_a_pos >= 30 or player_b_pos >= 30:
                break

            # 잡기 로직 (나머지 코드는 그대로 유지됨)
            if player_a_pos == player_b_pos and player_a_pos != 0:
                print(f"{current_player}가 상대방의 말을 잡았습니다!")
                if current_player == 'A':
                    player_b_pos = 0
                else:
                    player_a_pos = 0
                print(f"A의 위치: {player_a_pos}, B의 위치: {player_b_pos}")

            current_player = 'B' if current_player == 'A' else 'A'

    if player_a_pos >= 30:
        print("플레이어 A가 이겼습니다! 축하합니다!")
    else:
        print("플레이어 B가 이겼습니다! 축하합니다!")

if __name__ == "__main__":
    main() 
```
