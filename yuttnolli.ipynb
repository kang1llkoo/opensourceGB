```python
import random

def yut_throw():
    result = random.choices(["앞", "뒤"], k=4)
    result_str = ' '.join(result)
    if result.count("뒤") == 1:
        return result_str + " (도)"
    elif result.count("뒤") == 2:
        return result_str + " (개)"
    elif result.count("뒤") == 3:
        return result_str + " (걸)"
    elif result.count("뒤") == 4:
        return result_str + " (윷)"
    else:
        return result_str + " (모)"

def main():
    player_a_pos = 0
    player_b_pos = 0

    while player_a_pos < 29 and player_b_pos < 29:
        input("A의 차례입니다. 가위바위보를 시작하려면 엔터를 누르세요.")
        a_choice = random.choice(["가위", "바위", "보"])
        print(f"A의 선택: {a_choice}")

        input("B의 차례입니다. 가위바위보를 시작하려면 엔터를 누르세요.")
        b_choice = random.choice(["가위", "바위", "보"])
        print(f"B의 선택: {b_choice}")

        if a_choice == b_choice:
            print("무승부! 다시 가위바위보를 진행합니다.")
            continue
        elif (
            (a_choice == "가위" and b_choice == "보") or
            (a_choice == "바위" and b_choice == "가위") or
            (a_choice == "보" and b_choice == "바위")
        ):
            print("A가 이겼습니다!")
            player_a_pos = 0
        else:
            print("B가 이겼습니다!")
            player_b_pos = 0

        while player_a_pos < 29 and player_b_pos < 29:
            while True:
                input("A의 차례입니다. 윷을 던지려면 엔터를 누르세요.")
                yut_result = yut_throw()
                print(f"A의 윷 결과: {yut_result}")
                if '윷' in yut_result or '모' in yut_result:
                    print("윷이나 모가 나왔습니다. 한 번 더 던지십시오.")
                else:
                    break

            move_steps = yut_result.count("도") + 2 * yut_result.count("개") + 3 * yut_result.count("걸") + 4 * yut_result.count("윷") + 5 * yut_result.count("모")
            player_a_pos += move_steps
            if player_a_pos in [5, 20, 10]:
                player_a_pos = 20
            print(f"A의 위치: {player_a_pos}")

            if player_a_pos == player_b_pos:
                player_b_pos = 0
                print("A가 B의 말을 잡았습니다!")
                break

            while True:
                input("B의 차례입니다. 윷을 던지려면 엔터를 누르세요.")
                yut_result = yut_throw()
                print(f"B의 윷 결과: {yut_result}")
                if '윷' in yut_result or '모' in yut_result:
                    print("윷이나 모가 나왔습니다. 한 번 더 던지십시오.")
                else:
                    break

            move_steps = yut_result.count("도") + 2 * yut_result.count("개") + 3 * yut_result.count("걸") + 4 * yut_result.count("윷") + 5 * yut_result.count("모")
            player_b_pos += move_steps
            if player_b_pos in [5, 10, 20]:
                player_b_pos = 20
            print(f"B의 위치: {player_b_pos}")

            if player_b_pos == player_a_pos:
                player_a_pos = 0
                print("B가 A의 말을 잡았습니다!")
                break

    if player_a_pos >= 29:
        print("플레이어 A가 이겼습니다! 축하합니다!")
    else:
        print("플레이어 B가 이겼습니다! 축하합니다!")

if __name__ == "__main__":
    main()
```
